```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
metadados = pd.read_csv('Metadados.csv','\s+,\s+',engine='python');
metadados
```

```python
nrProcesso = np.array(metadados['Nr. Processo'])
ano = []
sessao = []
for i in nrProcesso:
    aux = i.split('.')
    ano.append(aux)
ano = pd.DataFrame(ano)
metadados['Ano'] = ano[1]

cidades = ["Natal","Mossoró","Caicó","Assu","Pau dos Ferros","Ceará-Mirim"]
sessao = pd.DataFrame(ano[4])
sessao[sessao[4]=='8400'] = cidades[0]
sessao[sessao[4]=='8401'] = cidades[1]
sessao[sessao[4]=='8402'] = cidades[2]
sessao[sessao[4]=='8403'] = cidades[3]
sessao[sessao[4]=='8404'] = cidades[4]
sessao[sessao[4]=='8405'] = cidades[5]
metadados['Sessão'] = sessao
metadados.head()
```

```python
metadados['Procedente ou Improcedente(P/I)'][metadados['Procedente ou Improcedente(P/I)']=='N,'] = '-'
metadados['Se teve tutela ,'][metadados['Se teve tutela ,']=='S,'] = 'S'
metadados['Se teve tutela ,'][metadados['Se teve tutela ,']=='N,'] = 'N'
metadados.describe()
```

```python
tutelados = metadados[metadados['Se teve tutela ,'] == 'S']
a = pd.DataFrame(tutelados['Assunto'].value_counts())
a.sort_index()
```

```python
tutelados_P = tutelados[tutelados['Procedente ou Improcedente(P/I)']=='P']
assuntos_tutelados_P = tutelados_P['Assunto'].value_counts()
b = pd.DataFrame(assuntos_tutelados_P)
b.sort_index()
```

```python
c = 100*b/a
c
```

```python
c = c.dropna(axis=0, how='all')
c
```

```python
saida = c[c['Assunto']>=50].sort_values(by='Assunto')
saida
```
